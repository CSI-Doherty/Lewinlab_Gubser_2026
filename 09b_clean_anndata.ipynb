```python
import anndata as ad
```

```python
adata = ad.read_h5ad("./data/converted_seurat_object.h5ad")
```

```python
# Remove tetramer cells
adata = adata[~adata.obs['sample'].str.contains('PBMC')]
```

```python
adata.obs['expand_response'] = adata.obs['expanded'].astype(str) + "_"+ adata.obs['time'].astype(str)
```

```python
# Create the inverse replacements dictionary
replacements = {
    "Donor 38": "P1",
    "Donor 41": "P2",
    "Donor 48": "P3",
    "Donor 49": "P4",
    "Donor 54": "P5",
    "Donor 56": "P6",
    "Donor 62": "P7",
    "Donor 64": "P8"
}

# Replace the values in the patient_name column
adata.obs['patient_name'] = adata.obs['sample'].replace(replacements)
```

```python
adata = adata[~adata.obs['patient_name'].isin(['P3','P5'])] #remove patient that have only baseline info
```

```python
adata.layers['raw_count'] = adata.X
```

```python
adata = adata[:,~adata.var.index.str.startswith('RPL')]
adata = adata[:,~adata.var.index.str.startswith('RPS')]
```

```python
adata.write_h5ad("./data/adata_cleaned.h5ad")
```
